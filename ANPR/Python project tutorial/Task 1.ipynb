### Write a function to draw a circle of radius 30 of blue color when left button of mouse is clicked down amd draw a circle of radius 30 of red color when right button of mouse is clicked down on a 512 by 512 black screen

```python
image_blank=np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(image_blank)
```

```python
image_blank=np.zeros(shape=(512,512,3),dtype=np.uint8)
def click_event(event,x,y,flags,param):
    if event==cv.EVENT_LBUTTONDOWN:
        strXY=str(x)+","+str(y)
        circle=cv.circle(image_blank,(x,y),30,(0,0,255),-1)
        plt.imshow(circle)
        cv.imshow('Coordinate Image',image_blank)
    if event==cv.EVENT_RBUTTONDOWN:
        strXY=str(x)+","+str(y)
        circle=cv.circle(image_blank,(x,y),30,(255,0,0),-1)
        plt.imshow(circle)
        cv.imshow('Coordinate Image',image_blank)
        
cv.imshow('Coordinate Image',image_blank)
cv.setMouseCallback('Coordinate Image',click_event)
cv.waitKey(0)
cv.destroyAllWindows()
```
